```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

## 1. Define initial population

```python
def g1(x):
    n = len(x)
    x = np.array(x)
    return 1 + 9 / (n - 1) * (x.sum() - x[0])
```

```python
def f_values(x):
    f1 = x[0]

    g = g1(x)
    f2 = g * (1 - math.sqrt(f1 / g))

    return f1, f2
```

## 2. Non-dominated sort

```python
def nd_sort(population):
    gen = population.copy()
    f1 = [f_values(individual)[0] for individual in gen]
    f2 = [f_values(individual)[1] for individual in gen]
    
    fronts = []
    while len(f1) > 0:
        front_i = []
        for i in range(len(f1)):
            if not any(((f1[j] <= f1[i] and f2[j] <= f2[i]) and (f1[j] < f1[i] or f2[j] < f2[i])) for j in range(len(f1))):
                front_i.append(gen[i])
        fronts.append(front_i)
        
        for individual in front_i:
            f1.remove(f_values(individual)[0])
            f2.remove(f_values(individual)[1])
            gen.remove(individual)
            
    return fronts
```

## 3. Crowding distance calculation

```python
def crowd_distance(individual, population):
    f1 = [f_values(individual)[0] for individual in population]
    f2 = [f_values(individual)[1] for individual in population]
    f1_value = f_values(individual)[0]
    f2_value = f_values(individual)[1]
    for front in nd_sort(population):
        if individual in front:
            if f1_value == max(f_values(ind)[0] for ind in front) or f2_value == max(f_values(ind)[1] for ind in front):
                return math.inf
            else:
                i = sorted(front).index(individual)
                return (front[i + 1][0] - front[i - 1][0])/(max(f1) - min(f1)) + (front[i + 1][1] - front[i - 1][1]) /(max(f2) - min(f2))
```

## 4. Selection and making offsprings

```python
def binary_tour(population):
    good_individuals = []

    while len(good_individuals) != 2:
        i = np.random.random_integers(0, len(population) - 1)
        j = None
        while j!=i:
            j = np.random.random_integers(0, len(population) - 1)
        
        rank_i = 0
        rank_j = 0
        for k in range(len(nd_sort(population))):
            if population[i] in nd_sort(population)[k]:
                rank_i = k+1
            elif population[j] in nd_sort(population)[k]:
                rank_j = k+1
        
        if (rank_i < rank_j) or (rank_i == rank_j and crowd_distance(population[i], population) > crowd_distance(population[j], population)):
            good_individuals.append(population[i])
        else:
            good_individuals.append(population[j])

    return good_individuals
```

```python
def crossover(parent1, parent2):
    size = min(len(parent1), len(parent2))
    cut_point = random.randint(1, size-1)
    parent1[cut_point:], parent2[cut_point:] = parent2[cut_point:], parent1[cut_point:]
    return parent1, parent2
```

```python
def mutation(child):
    mutation_rates = np.random.rand(len(child))
    mutated = np.random.rand(len(child))
    return np.where(mutation_rates < 0.05, mutated, child)
```

```python
def children(population):
    children_list = population.copy()  # children
    while len(children_list) != 2*len(population):  # create n children
        parents = binary_tour(population)  
        child1, child2 = crossover(parents[0], parents[1])
        child1 = mutation(child1)
        child2 = mutation(child2)
        if list(child1) not in children_list:
            children_list.append(list(child1))
        elif list(child2) not in children_list:
            children_list.append(list(child2))

    return children_list
```

```python
def new_generation(population):
    n = len(population)
    new_population = children(population)
    Pt1 = []
    for front in nd_sort(new_population):
        if len(Pt1) + len(front) <= n:
            Pt1 += front  # we copy first few fronts
        else:
            while len(Pt1) != n:
                cd = -math.inf
                add_to_pop = None
                for individual in front:
                    cdi = crowd_distance(individual, new_population)
                    if cdi > cd:
                        cd = cdi
                        add_to_pop = individual
                Pt1.append(add_to_pop)
                front.remove(add_to_pop)
    return Pt1
```

```python
def nsga2(initial_gen, max_gen):

    generation = new_generation(initial_gen)
    gen = 1
    
    while gen < max_gen:  # until we exceed max generation number, evolve
        generation = new_generation(generation)
        gen += 1

    pareto_x = [f_values(individual)[0] for individual in generation]
    pareto_y = [f_values(individual)[1] for individual in generation]
    plt.scatter(pareto_x, pareto_y)
    return generation
```
